# Sigmoid Function:

**Sigmoid Function** is a special case of **Logistic Function**. A **Logistic Function** is a $S$ shaped curve.
Equation is given by,

$$f(x) = \frac {L}{1 + e^{-k(x-x_0)}}$$

Where:

$L$ = Curve's Maximum Value

$x_0$ = Midpoint of the x-value

$K$ = Steepness of the curve


If we replace $L = 1, x_0 = 0, K = 1 $

we get the equation for the **Sigmoid curve**

$f(x) = \frac {1}{1 + e^{-(x)}}$

Reference:[Wikipedia](https://en.wikipedia.org/wiki/Logistic_function)

## Sigmoid Function Visualization:

Let us try to visualize the sigmoid function

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

In practice, due to the nature of the exponential function $\mathrm{e}^{-x}$, it is often sufficient to compute the standard logistic function for $x$ over a small range of real numbers such as a range contained in [âˆ’6, +6] as it quickly converges very close to its saturation values of 0 and 1. Hence we take the value $x = [-6,+6]$ in the below code.

```python
x = np.arange(-6,6,0.2)
y = np.linspace(0,len(x),len(x))
x.shape, y.shape
```

Without the sigmod function, our curve looks like a line as shown below.

```python
plt.plot(x,y)
```

```python
def sigmoid(x,L=1,K=1,M=0):
    a = []
    for item in x:
        #K = Steepness
        #M = X0 = Midpoint value of x
        #L = Curves Max value
        alpha = K*(item - M)
        a.append(L/(1+ math.exp(-alpha)))
    return a
```

Let us see how the curve looks for default values for our **Sigmoid Function**.

```python
#x = np.arange(-10,10,0.2)
y = sigmoid(x)
plt.plot(x,y)
```

Here is the visual of the curve for different values of $L$.

```python
y1 = sigmoid(x,2)
y2 = sigmoid(x,5)
y3 = sigmoid(x,10)
y4 = sigmoid(x,15)
plt.plot(x, y1, 'r', alpha=0.4)
plt.plot(x, y2, 'g',alpha=0.5)
plt.plot(x, y3, 'b',alpha=0.6)
plt.plot(x, y4, 'k',alpha=0.8)
plt.legend(['L=2','L=5','L=10','L=15'])
plt.show()
```

As you can observe in the above figure that, by varying the value of L, y-axis value varies.

Similarly, now let us try to visualize, how the changes in $K$ and $M$ affects the steepness of the curve.

```python
y1 = sigmoid(x, K=0.5, M=0.1)
y2 = sigmoid(x, K=0.3, M=0.01)
y3 = sigmoid(x, K=1.5, M=2)

plt.plot(x, y, 'k--', alpha=0.8)
plt.plot(x, y1, 'ro', alpha=0.6)
plt.plot(x, y2, 'g+', alpha=0.5)
plt.plot(x, y3, 'b*', alpha=0.4)
plt.legend(['K=1,M=0','K=0.5,M=0.1', 'K=0.3,M=0.01', 'K=1.5,M=2'])
plt.show()
```

Keeping $M=0$ and varying $K$, the steepness parameter

```python
y1 = sigmoid(x, K=0.5)
y2 = sigmoid(x, K=0.3)
y3 = sigmoid(x, K=1.5)

plt.plot(x, y, 'k--', alpha=0.8)
plt.plot(x, y1, 'ro', alpha=0.6)
plt.plot(x, y2, 'g+', alpha=0.5)
plt.plot(x, y3, 'b*', alpha=0.4)
plt.legend(['K=1, M=0','K=0.5,M=0', 'K=0.3,M=0', 'K=1.5,M=0'])
plt.show()
```
